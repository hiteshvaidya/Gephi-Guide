### I Start a new project
1. open the folder containing gephi eg. gephi-0.9.2
2. right click and select open in terminal\
3. In the terminal type, `bin/gephi`
4. The gephi window pops-up and select the option **New Project**
  ![](images/1.png)
<br>
<br>
5. Select *File* option in the Task bar and select either of the options to open a dataset.
  ![](images/2.png)
<br>
<br>
6. Select one *or* multiple files in the form of *.csv* or *.xlsx* 
  ![](images/3.png)
<br>
<br>
7. Make sure that the option **Adjacency list** is selected in **import as** menu.
  ![](images/4.png)
<br>
<br>
8. Keep on selecting **Next** *or* **Finish** until following window pops-up
  ![](images/5.png)
<br>
<br>
9. Make sure the undirected option is selected in the dropdown menu. Further the options in the radio button have following meaning:
   - **Create multiple workspaces** - It will create a new workspace for each new .csv or .xlsx file
   - **Merge into new workspace** - All the .csv or .xlsx files would be mereged into one new workspace
   - **Append to existing workspace** - All the .csv or .xlsx files would be merged and added to the existing workspace. This is option is particularly useful when there is already some graph opened in the existing workspace and you want to add another graph to it.
   - Based on your choice, you could select either of second or third radio button and then click on OK

### III Options on the left side of graph window
**Configure option**: The configure option on left-top determines the range of selection of mouse pointer. User may increase or decrease the radius of selection by dragging the pointer. 
![](images/10.png)
<br>
To operate other features (in the column below configure button) on multiple nodes, you can use the aforementioned feature of **configure**. One such example of moving/dragging multiple nodes is mentioned below.
<br>
All the options below configure button are self-explanatory. Just hover over those buttons and the explaination would pop up.

### How to do operations on multiple nodes:
   - To drag multiple nodes in Gephi, use the cylinder tool, which is part of the Dragging tool (Fist icon in the vertical bar).
     - While the Dragging tool is selected, in the top left of the Graph window, next to the word "Dragging," click the blue word (Configure) as shown above.
     - Set the diameter to something greater than one, and you'll see a grey disc that allows you to drag all nodes that are within the disc.
     - To "close" the tool, i.e. restore it to 1-pixel Dragging, Configure the diameter back to 1.
   - Similarly, you can do various other operations.

4. Find common friends between 2 nodes and extract them in a separate file
   - **Installation** : The Scripting Plugin can be installed through the Gephi's Tools/Plug-ins menu:
     - Go to the Tools > Plug-ins menu on Gephi's user interface.
     - Open the Available Plugins tab.
     - Click the "Reload Catalog" button, to make sure you have an up-to-date catalog.
     - Select the "Scripting Plugin" on the list and click the Install button.
     - After installation, the plugin can be accessed through the left slide on Gephi's UI (as shown on figure) or through the Window > Console menu.
         
     ![](images/22.png)
     
   <br>
   - **Usage** :  
      - Open **Data Table** window in gephi. If this window is not open already, click 'Window' in the task bar at the top. Then click on 'Data Table' option and the Data Table window would open.
      - Cick on the **Node** tab in the Data Table window as shown below in the image.
      ![](images/23.png)
      <br>
      - press 'ctrl' and select names of all the nodes whose intersection is to be found. 
      - Then copy them by typing 'ctrl' and 'c'. Open a new text file in notepad and paste these names there by typing 'ctrl' and 'v'.
      - **Attention!**: please mention the colour name that you want to give to the common nodes. This program will only detect _black, <font color='blue'>blue</font>, <font color='cyan'>cyan</font>, <font color='darkgray'>darkgray</font>, <font color='gray'>gray</font>, <font color='green'>green</font>, <font color='yellow'>yellow</font>, <font color='lightgray'>lightgray</font>, <font color='magenta'>magenta</font>, <font color='orange'>orange</font>, <font color='pink'>pink</font>, <font color='red'>red</font>, white._
      - If you wish to add any colour in this list, you can edit the `common.py` file yourself. Between line number 35 to line number 58, you can add two lines of code for the colour of your choice. Just copy the two lines of any other colour and replace it with your colour and paste it in the given range.
      - save this file with the name, **'nodes.txt'** in the gephi folder.
      - Make sure that the file **common.py** stays in the same folder of gephi.
      - Now in console window in gephi, type command `execfile('common.py')` and press 'enter'
      - If the nodes/clusters have anything in common then the names would be flashed and their colour would be changed. Otherwise, a message will be printed saying that there is nothing common between the given nodes.
